Gradient descent is an abstraction of Euler's Method, using multiple variables instead of single variables. Let me first show Euler's method:

```python
x = []
for i in range(-100,100,1):
    x.append(i/100)
y = np.square(x)
mpl.pyplot.scatter(x,y)
mpl.pyplot.show()
```

Lets do Euler's Method on c(w), our "cost function" that is just w^2. We want to find the minimum of c(w). Each optimization step, we will descend a bit according to the derivative. Basically, we will be doing w -= a * c'(w), where a is our learning rate. The graph shows the plot of our cost function as the blue line and each descent step as a red dot. After 10 iterations, we are close to the local/global min.

```python
def c(w):
    return w*w
w = 1
gdx=[]
gdy=[]
a=0.1

line = plt.plot(x,y)

for i in range(10):
    w -= a * 2*w
    gdx.append(w)
    gdy.append(c(w))
    plt.show()
    mpl.pyplot.plot(x,y,'b-',gdx,gdy,'ro')
    time.sleep(0.1)
    clear_output(wait=True)
```

Now, lets start doing linear descent: Basically, w is now a vector, consisting of weights for our prediction function theta(x)= w1* x + w2, and our cost function is now the MSE between our predictions and actual y values.

```python
'''generate some test data. Should give us the weights [3,4] since the equation is 3x+4'''
x = np.random.randn(100,1)
y = 4 +3 * x + np.random.randn(100,1)
mpl.pyplot.scatter(x,y)
mpl.pyplot.show()
```

```python
def cal_cost(theta,x,y):
    '''
    theta is the coefficients in our "model", x is all our inputs, y is our output.
    '''
    predictions = x.dot(theta)
    m = len(y)
    cost= (1/(2*m)) * np.sum(np.square(predictions - y))
    
    return cost
```

```python
def gradient_descent(x,y,theta,learningrate,iterations):
    m = len(y)
    cost_hist = np.zeros(iterations)
    theta_hist = np.zeros((iterations,2))
    
    for i in range(iterations):
        theta = theta - (1/m) * learningrate * (x.T.dot( x.dot(theta) - y ))
        theta_hist[i,:] = theta.T
        cost_hist[i] = cal_cost(theta,x,y)
    
    return theta, cost_hist, theta_hist
```

```python
lr = 0.01
n_iter = 1000

theta = np.random.randn(2,1)
x_b = np.c_[np.ones((len(x),1)),x]
theta, c_h, t_h = gradient_descent(x_b, y, theta, 0.01, 500)
print(theta)
```

We can see that our theta is close to what it should be. Now, lets plot cost history against iterations. Remember that cost history here plays the same role as the values of the cost function from Euler's Method example.

```python
mpl.pyplot.scatter(range(len(c_h)),c_h)
```
